### Retorno da taxa Simples

Calcule o retorno simples das ações do dica de hoje

$$
\frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1
$$

```python
egie['ret_simples'] = (egie['Adj Close'] / egie['Adj Close'].shift(1)) - 1
print(egie['ret_simples'])
```
