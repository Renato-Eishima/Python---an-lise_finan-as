```python
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
egie = yf.download('egie3.SA', start = '2017-01-01')
itsa = yf.download('ITSA4.SA', start = '2017-01-01')
sula = yf.download('SULA11.SA', start = '2017-01-01')
abev = yf.download('ABEV3.SA', start = '2017-01-01')
lren = yf.download('LREN3.SA', start = '2017-01-01')
slce = yf.download('SLCE3.SA', start = '2017-01-01')
bbas = yf.download('BBAS3.SA', start = '2017-01-01')
csan = yf.download('CSAN3.SA', start = '2017-01-01')
b3sa = yf.download('B3SA3.SA', start = '2017-01-01')
```

## Retorno Logarítmico

$$
ln(\frac{P_t}{P_{t-1}})
$$

```python
egie['ret_logaritmico'] = np.log(egie['Adj Close'] / egie['Adj Close'].shift(1))
print(egie['ret_logaritmico'])
```
